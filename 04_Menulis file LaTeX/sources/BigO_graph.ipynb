# Program O(n^2)

> mencetak grafik $O(n^2)$ menggunakan matplotlib kemudian menyimpannya ke dalam bentuk `.png` agar bisa kita gunakan dalam dokumen `LaTeX` kita.

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
 
# Creating vectors X and Y
x = np.linspace(0, 10, 100)
y = x ** 2
 
fig = plt.figure(figsize = (10, 5))
# Create the plot
plt.title('Grafik Kompleksitas')
plt.xlabel('$n$')
plt.ylabel('$O(n^2)$')
plt.plot(x, y)
plt.legend(['$O(n^2)$'], loc ="upper right")
plt.savefig('onn.png') # menyimpan gambar plot

# Show the plot
plt.show()
```
